### Sistema Massa-Mola

```python
import sympy
from sympy.abc import m, b, k, t, s # mass, damping, spring constant, time, Laplace variable
from sympy import Function, dsolve, Eq, integrate, laplace_transform, inverse_laplace_transform, DiracDelta, symbols

# Define the dependent variable (position x as a function of time t)
x = Function('x')(t)

# Define the external force (assuming an impulse for transfer function derivation, or general F(t))
F = Function('F')(t)

# 1. Differential Equation for Mass-Spring-Damper System
# Newton's Second Law: m*a = Sum of Forces
# Forces: -k*x (spring), -b*x_dot (damper), +F (external)
x_dot = sympy.diff(x, t)
x_double_dot = sympy.diff(x, t, 2)

diff_eq = Eq(m * x_double_dot + b * x_dot + k * x, F)

print("1. Differential Equation of the Mass-Spring-Damper System:")
sympy.pprint(diff_eq)
print("\n")

# 2. Transfer Function Derivation (assuming zero initial conditions)
# Apply Laplace Transform to the differential equation
# L{d^2x/dt^2} = s^2 X(s)
# L{dx/dt} = s X(s)
# L{x} = X(s)
# L{F(t)} = F(s)

X_s = Function('X')(s)
F_s = Function('F')(s)

laplace_diff_eq = Eq(m * s**2 * X_s + b * s * X_s + k * X_s, F_s)

print("2. Laplace Transformed Equation (zero initial conditions):")
sympy.pprint(laplace_diff_eq)
print("\n")

# Solve for X(s)/F(s) to get the Transfer Function H(s)
H_s = sympy.solve(laplace_diff_eq, X_s)[0] / F_s

print("3. Transfer Function H(s) = X(s)/F(s):")
sympy.pprint(H_s)
print("\n")

# 4. State-Space Representation
# Let's define state variables:
# x1 = x (position)
# x2 = x_dot (velocity)
# Then:
# x1_dot = x2
# From the differential equation: m*x2_dot + b*x2 + k*x1 = F
# x2_dot = (-b/m)*x2 - (k/m)*x1 + (1/m)*F

# Define state variables as symbols for matrices
# State vector: X = [x; x_dot]
# Input: u = F
# Output: y = x

A = sympy.Matrix([[0, 1], [-k/m, -b/m]])
B = sympy.Matrix([[0], [1/m]])
C = sympy.Matrix([[1, 0]])
D = sympy.Matrix([[0]])

print("4. State-Space Representation:")
print("   dx/dt = A * x + B * u")
print("   y     = C * x + D * u")
print("\n   A matrix:")
sympy.pprint(A)
print("\n   B matrix:")
sympy.pprint(B)
print("\n   C matrix:")
sympy.pprint(C)
print("\n   D matrix:")
sympy.pprint(D)
```
